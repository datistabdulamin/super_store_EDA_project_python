```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("Superstore_USA.xlsx")
```

```python
df.head()
```

```python
# checking mission values:
```

```python
df.isnull().sum()
```

```python
# in, [Product Base Margin] the [72] values is missing. 
```

```python
# fill the missing values with the mean of the Coloum.
```

```python
df["Product Base Margin"].fillna(df["Product Base Margin"].mean() ,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

# Order Priority

```python
df["Order Priority"].unique()
```

```python
df["Order Priority"].value_counts()
```

```python
# make graph on Oder Priority:
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Order Priority",data=df)
plt.title("Count of Order Priority")
plt.savefig("Count of Order Priority.jpg")
plt.show()
```

# Ship Mode

```python
df["Ship Mode"].unique()
```

```python
df["Ship Mode"].value_counts()
```

```python
# make pie chart for shipping Mode:
name = df["Ship Mode"].value_counts().index
value = df["Ship Mode"].value_counts().values
```
